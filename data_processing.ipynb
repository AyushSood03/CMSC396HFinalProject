```python
import pandas as pd
import numpy as np
```

```python
amd_df = pd.read_csv("drive/MyDrive/CMSC396H/agricultural_monthly_data.csv")
cpi_df = pd.read_csv("drive/MyDrive/CMSC396H/monthly_CPI_20yrs.csv")
w_df = pd.read_csv("drive/MyDrive/CMSC396H/weatherdata.csv")
```

```python
amd_df
```

```python
amd_df = amd_df.sort_values(by=['date'], ascending=True)
amd_df
```

```python
cpi_df
```

```python
w_df
```

```python
amd_df = amd_df[['date', 'TICKER', 'RET']]
amd_df
```

```python
cpi_df = cpi_df.rename(columns={'caldt':'date', 'b30ret':'b30ret', 'b20ret':'b20ret', 'b10ret':'b10ret', 'cpiret':'cpiret'})
cpi_df
```

```python
amd_cpi_df = pd.merge(amd_df, cpi_df, on='date')
amd_cpi_df
```

```python
amd_cpi_df = amd_cpi_df[['date', 'TICKER', 'RET', 'cpiret']]
amd_cpi_df
```

```python
amd_cpi_df = amd_cpi_df.loc[amd_cpi_df['TICKER'].isin(['ADM', 'BG', 'TSN'])]
amd_cpi_df
```

```python
w_df = w_df.rename(columns={'YYYY-MM':'YYYY-MM', 'Fort Worth,TX': 'Fort Worth, TX', 'Audbon, IA': 'Audbon, IA', 'Neosho ,MO': 'Neosho, MO'})
w_df = w_df[['YYYY-MM', 'Fort Worth, TX', 'Audbon, IA', 'Neosho, MO']]
w_df
```

```python
TXavg = w_df.loc[:, 'Fort Worth, TX'].mean()
IAavg = w_df.loc[:, 'Audbon, IA'].mean()
MOavg = w_df.loc[:, 'Neosho, MO'].mean()

for index, row in enumerate(w_df['Fort Worth, TX']):
  if not pd.notnull(row):
    w_df['Fort Worth, TX'][index] = TXavg

for index, row in enumerate(w_df['Audbon, IA']):
  if not pd.notnull(row):
    w_df['Audbon, IA'][index] = IAavg

for index, row in enumerate(w_df['Neosho, MO']):
  if not pd.notnull(row):
    w_df['Neosho, MO'][index] = MOavg
```

```python
amd_cpi_df['date'] = amd_cpi_df['date'].str[:7]
amd_cpi_df
```

```python
amd_cpi_df = amd_cpi_df.rename(columns={'date': 'YYYY-MM', 'TICKER': 'TICKER', 'RET': 'RET', 'cpiret': 'cpiret'})
amd_cpi_df
```

```python
amd_cpi_w_df = pd.merge(amd_cpi_df, w_df, on='YYYY-MM')
amd_cpi_w_df
```

```python
for index, row in enumerate(amd_cpi_w_df['RET']):
  if amd_cpi_w_df['RET'][index] == 'C':
    amd_cpi_w_df['RET'][index] = amd_cpi_w_df['RET'][index + 1]
for index, row in enumerate(amd_cpi_w_df['cpiret']):
  if amd_cpi_w_df['cpiret'][index] == 'C':
    print('cpiret issue')
amd_cpi_w_df
```

```python
amd_cpi_w_df['Real RET'] = amd_cpi_w_df['RET'].astype(np.float64) - amd_cpi_w_df['cpiret'].astype(np.float64)
amd_cpi_w_df
```

```python
amd_cpi_w_df.to_csv('returns_and_weather.csv')
```
